```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df_Titanic = pd.read_csv('Titanic-Dataset.csv')
df_Titanic
```

```python
df_Titanic.info()
```

```python
df_Titanic.describe()
```

```python
df_Titanic.describe().sum()
```

```python
df_Titanic.isna()
```

```python
df_Titanic.isna().sum()
```

```python
df_Titanic['Age'].fillna(method = 'ffill', inplace = True)
df_Titanic['Fare'].fillna(method = 'ffill', inplace = True)
df_Titanic
```

```python
df_Titanic.isna().sum()
```

```python
df_Titanic = df_Titanic.sort_values(ascending=False, by=['Survived'])
df_Titanic
```

```python
pivot_table = pd.pivot_table(df_Titanic, values=['PassengerId','Survived','Age'],index = 'Sex',aggfunc='mean')
pivot_table
```

```python
pivot_table = pivot_table.reset_index()
pivot_table
```

```python
df_Titanic.drop(['Pclass','SibSp','Parch','Ticket','Fare','Cabin','Embarked'],axis=1,inplace= True)
df_Titanic
```

```python
df_tat = df_Titanic
```

```python
Total_Survived = pivot_table['Survived'].sum()
plt.figure(figsize=(10, 5))
bars = plt.bar(pivot_table['Sex'], pivot_table['Survived'], color=plt.cm.PRGn(np.linspace(0, 1, len(pivot_table))))
plt.xlabel('Sex',fontsize=20, fontweight='bold')
plt.ylabel('Survived (%)',fontsize=20, fontweight='bold')
plt.title("Survival percentage(%) as per Sex",fontsize =20, fontweight='bold')
# Adding data labels with percentages
for bar in bars:
    yval = bar.get_height()
    percent = (yval / Total_Survived) * 100
    plt.text(bar.get_x() + bar.get_width() / 2, yval, f'{percent:.2f}%', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
Total_Survived = pivot_table['Survived'].sum()
pivot_table['Percentage'] = (pivot_table['Survived'] / Total_Survived) * 100
plt.figure(figsize=(10, 5))
plt.pie(pivot_table['Survived'], labels=pivot_table['Sex'], autopct='%.2f%%', startangle=90, colors=plt.cm.Set1(np.linspace(0, 1, len(pivot_table))))
plt.title("Survival percentage(%) as per Sex",fontsize =20, fontweight='bold')
plt.axis('equal')

plt.show()
```

```python
Age = pivot_table['Age'].sum()
plt.figure(figsize=(10, 5))
bars = plt.bar(pivot_table['Sex'], pivot_table['Age'], color=plt.cm.managua(np.linspace(0, 1, len(pivot_table))))
plt.xlabel('Sex',fontsize=20, fontweight='bold')
plt.ylabel('Age',fontsize=20, fontweight='bold')
plt.title("Age as per Sex",fontsize =20, fontweight='bold')
# Adding data labels with percentages
for bar in bars:
    yval = bar.get_height()
    percent = (yval / Age) * 100
    plt.text(bar.get_x() + bar.get_width() / 2, yval, f'{percent:.2f}%', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# To know about the numbers of the survived sex 
Sex_counts = df_tat['Sex'].value_counts()
plt.figure(figsize=(10, 6))
bars = plt.bar(Sex_counts.index, Sex_counts, color=plt.cm.PiYG(np.linspace(0, 1, len(Sex_counts))))
plt.xlabel('Sex', fontsize=20, fontweight='bold')
plt.ylabel('Count', fontsize=20, fontweight='bold')
plt.title('Survival numbers as per Sex', fontsize=20, fontweight='bold')

# Adding data labels
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2.0, yval, int(yval), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Create pie chart for Survival percentage as per sex
Sex_counts = df_tat['Sex'].value_counts()
plt.figure(figsize=(10, 6))
plt.pie(Sex_counts, labels=Sex_counts.index, autopct='%.2f%%', startangle=90, colors=plt.cm.managua(np.linspace(0,1,len(Sex_counts))))
plt.title('Survival percentage(%) as per sex', fontsize=20, fontweight='bold')
plt.show()
```
